```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Step-1: Generate 100 random values in the range [0,1]
data=np.random.rand(100)
```

```python
#Step-2: Label the first 50 values
labels=[]
for x in data [:50]:
    if x<=0.5:
        labels.append(1)   #Class 1
    else:
        labels.append(2)   #Class 2
```

```python
#Step-3: Prepare training data
train_data=data[:50].reshape(-1,1)
train_labels=np.array(labels)
```

```python
#Step-4: Prepare test data
test_data=data[50:].reshape(-1,1)
```

```python
#Step-5: Different values of k
k_values=[1,2,3,4,5,20,30]
```

```python
#Step-6: Perform KNN Classification
for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    #Train the model
    knn.fit(train_data,train_labels)
    #Predict class for remaining points
    predicted=knn.predict(test_data)
    print("\n Results for k=",k)
    print("Test value\t Predicted class")
    for value,label in zip(test_data.flatten(),predicted):
        print(f"{value:.3f}\t\t Class{label}")
```
